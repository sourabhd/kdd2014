```python
essays = pd.read_csv('../../dataset/essays.csv')
```

```python
#tv = TfidfVectorizer(tokenizer=lambda doc: doc, lowercase=False) 

lmtzr = WordNetLemmatizer()
stemmer = SnowballStemmer('english')

titleWordList = []

for title in essays['title'].fillna(''):
    
    # Create word list
    titleWords = title.encode('utf-16','replace')

    # tokenize 
    titleWords2 = wordpunct_tokenize(titleWords)

    # convert to lowercase
    titleWords3 = [w.lower() for w in titleWords2]

    # remove stop words
    titleWords4 = [ word for word in titleWords3 if word not in set(stopwords.words('english')) ]
    
    # lemmatize based on WordNet 
    titleWords5 = [ lmtzr.lemmatize(word) for word in titleWords4 ]
    
    # stem using snowball stemmer
    titleWords6 = [ stemmer.stem(word) for word in titleWords5 ]
    
    # remove punctuations
    titleWords7 = [ word.encode('utf-8').translate(None,string.punctuation) for word in titleWords6 ]
    
    # remove empty strings
    titleWords8 = [ word for word in titleWords7 if word <> '' ]
    
    titleWordList.append(titleWords8)
    
    #print titleWords
    #print titleWords2
    #print titleWords3
    #print titleWords4
    #print titleWords5
    #print titleWords6
    #print titleWords7
    #print titleWords8
    #print 

print len(titleWordList)
#tv = TfidfVectorizer()
#tv.fit(words)
#print tv.get_feature_names()
```
